 # Q7. 럭키 스트레이트

```python
score = input()
length = int(len(score)/2)

sum1 = 0
for i in range(length):
    sum1 += int(score[i])

sum2 = 0
for i in range(length,length*2):
    sum2 += int(score[i])

if sum1 == sum2:
    print("LUCKY")
else:
    print('READY')
```

 # Q8. 문자열 재정렬

```python
string = input()

letter = []
number = 0

for i in range(len(string)):
    try:
        number += int(string[i])
    except:
        letter.append(string[i])

letter.sort()
letter.append(str(number))

print("".join(letter))
```
